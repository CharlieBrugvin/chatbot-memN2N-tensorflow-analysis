# accuracy on train & test : Comparison over models

```python
# creating subplots
fig = plt.figure(figsize=(10,10))

ax = []
for i in range(6):
    ax_i = fig.add_subplot(321+i)
    ax.append(ax_i)
    
i = 0
for model_name, group in datas[['model_name', 'epoch', 'train_acc', 'val_acc']].groupby('model_name'):
    plot = group.plot(x='epoch', title = model_name, ax=ax[i])
    i += 1
    
plt.subplots_adjust(hspace=0.4)
```

# Computer calculation time

```python
df = datas[['model_name','epoch_timestamp']].groupby('model_name').agg({'epoch_timestamp': [min, max]}).reset_index()
```

```python
df.columns = ['model_name', 'start', 'end']
df['duration (minutes)'] = np.round((df.end - df.start) / 60.).astype(int)
```
