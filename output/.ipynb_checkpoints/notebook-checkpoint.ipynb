# Data preprocessing

```python
datas = pd.read_csv('batch_size.csv', delimiter=',',
            names=['model_id', 'task_id', 'batch_id', 'hops', 'nb_epoch',
                   'epoch', 'epoch_date', 'epoch_timestamp', 
                   'train_size_prct', 'train_size', 'val_size', 'test_size',
                   'train_acc', 'val_acc', 'test_acc'])
```

```python
df = datas[['model_id', 'epoch', 'test_acc']].copy()
df = pd.pivot_table(df, values='test_acc', columns=['model_id'], index='epoch')
```
